```python
import pandas as pd
```

```python
file_one = "northamerica_station_information (1).csv"
```

```python
file_two = "city_gasday_fcst (1).csv"
```

```python
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
```

```python
file_one_df.head()
```

```python
file_one_df.rename(columns = {"identifier": "Station"}, inplace = True)
```

```python
file_one_df.head()
```

```python
file_two_df = pd.read_csv(file_two, encoding="ISO-8859-1")
```

```python
file_two_df.head()
```

```python
merge_table = pd.merge(file_one_df, file_two_df, on="Station", how="right")
merge_table.head()
```

```python
del merge_table["HDD"]
```

```python
del merge_table["CDD"]
```

```python
merge_table.head()
```

```python
new_df = merge_table.loc[merge_table["country"] == "US",:]
```

```python
new_df.head()
```

```python
new_df1 = new_df.rename(columns = {"Station":"station_id","name":"station_name","Production Date":"production_date","Date":"forecast_date",
                         "Fcst Mn":"fcst_mn","Fcst Mx":"fcst_mx","Fcst Avg":"fcst_avg","Norm Mn":"norm_mn","Norm Max":"norm_mx"
                        }).head()
```

```python
new_df1.reset_index().rename(columns = {new_df1.index.name: "id"})
```

```python
Base = declarative_base()

class Forecast(Base):
    __tablename__ = 'forecast'
    station_id = Column(String, primary_key=True)
    
```
